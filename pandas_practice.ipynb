## Set-like Operations

```python
dfr = pd.DataFrame({
    "Names": ["Rizwan", "Sabiha", "Abdullah"],
    "Marks": [89, 56,32],
    "Age": [28, 29, 3],
})
dfs = pd.DataFrame({
    "Names": ["Rizwan", "Sabiha"],
    "Marks": [89, 56],
    "Age": [28, 29],})
```

Rows that appear in both dfx and dfy
(Intersection). If complete data of rows from both dataframes is same.

```python
pd.merge(dfr, dfs)
```

Rows that appear in either or both ydf and zdf
(Union).

```python
pd.merge(dfr, dfs, how = "outer")
```

It merges two DataFrames and adds a special column called _merge that shows the origin of each row.

```python
pd.merge(dfx, dfy, how='outer', 
         indicator=True)
```

.query('_merge == "left_only"') filters the merged DataFrame and keeps only those rows that exist only in the left DataFrame.

```python
pd.merge(dfx, dfy, how='outer', indicator=True).query('_merge == "left_only"')
```

.query('_merge == "right_only"') filters the merged DataFrame and keeps only those rows that exist only in the right DataFram

```python
pd.merge(dfx, dfy, how='outer', indicator=True).query('_merge == "right_only"')
```

.drop(columns=['_merge']) deletes the _merge column from the DataFrame.

```python
pd.merge(dfx, dfy, how='outer', indicator=True).query('_merge == "left_only"').drop(columns=['_merge'])
```

## Sub-plotting in Pandas

Subplotting in pandas means creating multiple plots in a single figure (grid layout) using a DataFrame, where each subplot shows a different column or relationship.

Instead of making one graph per figure, pandas lets you split one figure into many small plots.

**Line Plots**

```python
cols = [
    'BALANCE',
    'BALANCE_FREQUENCY',
    'PURCHASES',
    'ONEOFF_PURCHASES',
    'INSTALLMENTS_PURCHASES',
    'CASH_ADVANCE',
    'PURCHASES_FREQUENCY'
]

df7[cols].plot(
    subplots=True, title=['BALANCE',
    'BALANCE_FREQUENCY',
    'PURCHASES',
    'ONEOFF_PURCHASES',
    'INSTALLMENTS_PURCHASES',
    'CASH_ADVANCE',
    'PURCHASES_FREQUENCY'],
    layout=(4, 2),
    figsize=(16, 14),
    sharex=False,
    linewidth=2,
    #grid=True,
    legend=False
)
```

A Pandas Series is a one-dimensional labeled array in Python that can hold any type of data such as integers, floats, strings, or even Python objects.\
It is one of the core data structures in the Pandas library, designed for fast and efficient data analysis.

**Key Idea** :Very Important

All these are vectorized operations:

Work on entire Series at once\
Faster than Python loops\
Clean and pandas-friendly

1. s.str.count(pattern)

Counts how many times a pattern appears in each string.

```python
s = pd.Series(["apple", "banana", "apple pie"])
s.str.count("a")
```

2. s.str.get(index)

Gets character at a specific position.

```python
s = pd.Series(["apple", "banana"])
s.str.get(2)
```

3. s.str.join(sep)

Joins characters of each string with a separator.

```python
s = pd.Series(["apple", "banana"])
s.str.join("_")
```

4. s.str.title()

Capitalizes first letter or convert a word into a title case.

```python
s = pd.Series(["hello world", "data science"])
s.str.title()
```

s.str.len()

Returns length of each string.

```python
s = pd.Series(["hello world", "data science"])
s.str.len()
```

6. s.str.cat()

Concatenates strings (combine into one or multiple)

```python
s1 = pd.Series(["hello world", "data science"])
s2 = pd.Series(["hello rizwan", "Scientist"])
s.str.cat(s2)
```

7. s.str.partition(sep)

Splits into 3 parts: before, separator, after (first match only).

```python
s = pd.Series(["a-b-c", "x-y-z"])
s.str.partition("-")
```

Cuts part of each string.

## Display options

**Display.max_rows**:  is a Pandas option that determines the maximum number of rows shown when printing a DataFrame. In other words,\
it controls how many rows are displayed on the screen before Pandas starts truncating the output.

**Display.max_columns** is a Pandas option that controls the maximum number of columns displayed when showing a DataFrame. In other words,\
it decides how many columns will be shown on the screen before Pandas starts hiding the rest.

**Display.expand_frame_repr** is a Pandas option that controls whether a DataFrame is displayed across multiple lines or wrapped into a single line. In other words,\
it decides if wide DataFrames should be broken into multiple rows for readability or shown in a compact, wrapped format.

**Display.large_repr** is a Pandas option that controls how large DataFrames are displayed. In other words, it decides whether a large\
DataFrame should be fully shown or truncated (partially displayed) when printed.

**Display.precision** is a Pandas option that controls how many decimal places are shown when displaying numbers. In other words, it sets the number\
of digits after the decimal point for floating-point values in output.

**Display.max_colwidth** is a Pandas option that controls how much of a column’s content (especially text or strings) is displayed. In other words, it sets the\
maximum character width for displaying values in a DataFrame column; longer text may be truncated if it exceeds this limit.

**Display.max_info_columns** is a Pandas option that controls how many columns are shown in detail when using DataFrame.info(). In other words, it sets\
a limit: if the number of columns exceeds this value, Pandas may not display full column-level information in the info() output.

**Display.chop_threshold** is a Pandas option that makes very small numeric values appear as 0 when displayed. In other words, if numbers are smaller
than the set threshold, Pandas will “chop” them and show them as zero in the output (without changing the actual data).

**Display.colheader_justify** is a Pandas option that controls how DataFrame column headers are aligned. In other words, it determines whether
the column names are aligned to the left or right when displayed.
